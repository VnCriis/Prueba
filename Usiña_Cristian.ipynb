```python
import seaborn as sns
import matplotlib.pyplot as plt

resumen_df = df.groupby(['Make', 'Electric Vehicle Type']).size().unstack(fill_value=0).reset_index()

# Grafico de barras apiladas
plt.figure(figsize=(12, 8))
sns.barplot(x='Make', y='Battery Electric Vehicle (BEV)', data=resumen_df, color='skyblue', label='VehÃ­culos ElÃ©ctricos')
sns.barplot(x='Make', y='Plug-in Hybrid Electric Vehicle (PHEV)', data=resumen_df, color='orange', bottom=resumen_df['Battery Electric Vehicle (BEV)'], label='VehÃ­culos Plug-in HÃ­bridos')

plt.xlabel('Marca')
plt.ylabel('Cantidad de VehÃ­culos')
plt.title('ComparaciÃ³n de VehÃ­culos ElÃ©ctricos y Plug-in HÃ­bridos por Marca')
plt.xticks(rotation=45, ha='right')
plt.legend()

plt.show()
```
